```python
numbers = [1,2,3,4,5]

iter_numbers = iter(numbers)
while True: 
    try: 
        num = next(iter_numbers)
        print(num)
    except StopIteration:
        break
    
```

```python
class CountUp:
    def __init__(self,limit):
        self.limit = limit 
        self.current = 0 
    
    def __iter__(self):
        return self 
    
    def __next__(self):
        if self.current < self.limit : 
            self.current +=1
            return self.current
        else:
            raise StopIteration
counter = CountUp(8)
for num in counter:
    print(num)
```

### Generators

```python
def square(num):
    for i in range(num):
        yield i*i
for item in square(5):
    print(item)
```

```python
%%timeit
ans = [x*x for x in range(5)]
print(ans)
```

### list comprehension is faster than the generators 


```python
def logger(func):
    def wrapper(*args,**kwargs):
        print(f"Calling {func.__name__} with {args} {kwargs}")
        result = func(*args,**kwargs)
        print(f'Result = {result}')
        return result 
    return wrapper

@logger
def add(a,b):
    return a+b

add(3,4)
```

```python
import time 
def timer(func):
    def wrapper(*args,**kwargs):
        startTime = time.time()
        result = func()
        endTime= time.time()
        print(F"The time taken is {endTime-startTime}")
        return result 
    return wrapper 


@timer 
def sayHello():
    print("hello")

sayHello()
```

```python
class Even:
    def __init__(self,limit):
        self.limit = limit 
        self.current = 0 
        self.count = 0 
        
    def __iter__(self):
        return self 
    
    def __next__(self):
        if self.current < self.limit:
            self.current +=2
            self.count += 1 
            return self.current
        else :
            raise StopIteration
    
num = Even(5)
for item in num:
    print(item)
```

```python
def fibo(n):
    a,b = 0,1
    for _ in range(n):
        yield a
        a,b = b,a+b
        
for item in fibo(7):
    print(item)
```
