# Accessing DataBase in Python 

using sqlite

```python
import sqlite3

# connect or create a database 

conn=sqlite3.connect("exampleX.db")
cursor=conn.cursor()
```

```python
#Create a table 
cursor.execute("""
        CREATE TABLE IF NOT EXISTS students(
            id INTERGER PRIMARY KEY,
            name TEXT,
            age INTEGER,
            address TEXT
        )
               """)
```

```python
# Insert data 
cursor.execute("INSERT INTO students(name,age,address) VALUES(?,?,?)",
("sujal",5,"itahar"))
conn.commit()
```

```python
# select all query 

cursor.execute('SELECT * FROM students WHERE address LIKE "itahar" ')
rows=cursor.fetchall()
print(rows)
```

```python
conn.close()
```

```python
from sqlalchemy import create_engine

engine = create_engine("sqlite:///mydb.db",echo=True)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column,Integer,String

Base= declarative_base()

class Student(Base):
    __tablename__ = 'students'
    id=Column(Integer,primary_key=True)
    name=Column(String)
    age=Column(Integer)
    
#Create the table 
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
#Adding data 

student1=Student(name='Ram',age=25)
student2=Student(name='Suman',age=34)
session.add(student1)
session.add(student2)
session.commit()
```
