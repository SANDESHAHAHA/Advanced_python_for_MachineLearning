## Lambda function usecases

```python
data=[(1,'ram','biratnagar'),(2,'aman',"itahari"),(1,"hari","dharan")]
sorted_data=sorted(data,key=lambda x:x[1]) #sorting on the basis of the name 
print(sorted_data)
```

```python
#use_Case ii) filtering 

aList=[1,2,3,4,5,6,7,8,9]
even = list(filter(lambda x: x%2==0,aList))
print(even)
```

```python
#third use case 
aList=[1,2,3,4,5,6,7,8,9]
squares = list(map(lambda x: x**2,aList))
print(squares)
```

## OOP

```python
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print(f"{self.name} speaks!")
class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name)
        self.breed=breed
        
    def speak(self):
        super().speak()
        print(f"{self.name} barks ")
d1= Dog("lucy","shepherd")
d1.speak()
```

```python
class A:
    def show(self):
        print("A")
        
class B:
    def show(self):
        print("B")
        
class C(B,A):
    pass

c1=C()
c1.show()
```

## METHOD RESOLUTION ORDER (mro) is the way in which python call its constructor 

```python
C.mro()
```

Multi level Inheritance 

```python
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
        super().show()
class C(B):
    def show(self):
        print("C")
        super().show()
        
c1=C()
c1.show()
    
```

## Debugging using pdb not print 
n=next line 
s=step into function 
c = continue until next breakpoint
l=list current code lines 
p variable  =print value of variable 
q= quit debugger

```python
import pdb

x=[1,2,3,4,5]
pdb.set_trace() #program pauses here
y=x[5]
print(y)
```
