```python
from sqlalchemy import create_engine
engine=create_engine("sqlite:///books_db.db",echo=True)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column,Integer,String

Base =declarative_base()

class Book(Base):
    
    __tablename__ = "books"
    id=Column(Integer,primary_key=True)
    title=Column(String,nullable=False)
    author=Column(String,nullable=False)
    year=Column(String,nullable=False)
    
#create the tabel
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session=sessionmaker(bind=engine)
session=Session()
```

```python
#adding data
book_data1=Book(title="time travel",author="hg wells",year="1929")
book_data2=Book(title="the metamorphosis",author="franz kafka",year="1994")
book_data3=Book(title="the invincibleman",author="william shakespeare",year="1959")
session.add(book_data1)
session.add(book_data2)
session.add(book_data3)
```

```python
session.commit()
```

```python
#query by author name 
books_by_author=session.query(Book).filter(Book.author=="hg wells").all()
for book in books_by_author:
    print(book.title,book.year)
```

```python
#book to update 
session.query(Book).filter(Book.author =="hg wells").update({"year":"2000"})
session.commit()
updated_list=session.query(Book).all()
for book in updated_list:
    print(book.title,book.year)
```

```python
#delete a book 
session.query(Book).filter(Book.year == "1959").delete()
session.commit()

remaining_books=session.query(Book).all()
for  book in remaining_books:
    print(book.title,book.author,book.year)
    
```
