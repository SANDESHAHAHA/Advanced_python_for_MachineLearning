```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year 
        
    def display(self):
        print(f"Car is of {self.make} of model {self.model} made in year {self.year}")
        
c = Car('toyata','ABC',234)
c.display()
```

```python
class ElectricCar(Car):
    def __init__(self,make,model,year,battery_capacity):
        super().__init__(make,model,year)
        self.battery_capcity = battery_capacity
    def display_battery_info(self):
        print(f"The battery is {self.battery_capcity} KWH")
ev1 = ElectricCar("byd","dolphin",2025,45)
ev1.display()
ev1.display_battery_info()
```

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length 
        self.width = width 
    def area(self):
        print(f"The area is {self.length*self.width}")
    def perimeter(self):
        print(f"The perimeter is {2*(self.length+self.width)}")
        
r1 = Rectangle(2,3)
r1.perimeter()
r1.area()
```

```python
from math import pi
class Circle:
    def __init__(self,radius):
        self.__radius = radius 

    def set_radius(self,radius):
        self.__radius = radius 
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return pi*(self.__radius**2)
c1 = Circle(4)
print(f"The radius is {c1.get_radius()}")
c1.set_radius(5)
print(f"The radius is {c1.area()}")
```
