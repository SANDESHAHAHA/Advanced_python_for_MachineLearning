## Generators: 
A generator is a way of iterating using yield.
Generator calls are not like function calls .
Generators is used to save memory at run time execution .
Generators produces value only when needed .

```python
def squares(x):
    for i in range(x):
        # print (i*i)
        yield i*i
# squares(5)
x=(squares(5))
for num in x:
    print(num)
# print(type(squares(5)))
# for i in range(5):
#     print(i)
```

Decorators: 
it a function : that we get after wrapping a function by a wrapper 

```python
def mydecorator(func): 
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper 
```

```python
@mydecorator
def sayHello():
    print("Hello")
```

```python
sayHello()
```

```python
import time  
def timer(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end=time.time()
        print(f'Elapesed time = {end-start} seconds')
        return result 
    return wrapper
```

```python
@timer
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f
fact(100)
```
