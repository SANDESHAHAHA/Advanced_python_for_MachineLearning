# Matplotlib 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
fig = plt.figure()
print(fig.show())
```

```python
fig = plt.figure()
ax1= fig.add_subplot(2,2,1)
ax2= fig.add_subplot(2,2,2)
ax3= fig.add_subplot(2,2,3)
ax4= fig.add_subplot(2,2,4)
ax1.plot(np.random.standard_normal(50).cumsum(),color='black',linestyle='dashed')
ax2.hist(np.random.standard_normal(100),bins=20,color='black',alpha=0.3,edgecolor='black')
ax3.scatter(np.arange(30),np.arange(30)+3*np.random.standard_normal(30))
ax4.bar(x=['a','b','c'],height=[4,5,10]);
```

```python
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500),bins=50,color='black',alpha=0.5)
fig.subplots_adjust(wspace=0,hspace=0)
print(fig)
print(axes)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(1000).cumsum());
ticks = ax.set_xticks([0,250,400,600,800])
labels = ax.set_xticklabels(["one","two","three","four","five"],rotation=30,fontsize=10)
ax.set_title("My Plot")
ax.text(x=5,y=15,s="hello world",family="monospace",fontsize=10)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

X = np.arange(1,6)
y = X *2 
data = np.random.randn(100)
fig,ax =plt.subplots(2,2,figsize=(6,4))

ax[0,0].plot(X,y)
ax[0,0].set_title("Line plot")

ax[0,1].bar(X,y)
ax[0,1].set_title("Bar plot")

ax[1,0].hist(data)
ax[1,0].set_title("Histogram")

ax[1,1].scatter(X,y)
ax[1,1].set_title("ScatterPlot")

#fix spacing 
plt.tight_layout()
plt.show()
```
