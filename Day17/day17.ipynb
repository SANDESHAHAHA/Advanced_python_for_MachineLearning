```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
X = np.arange(1,10,0.2)
Y = np.square(X)
y_cube = np.power(X,3)
plt.plot(X,Y,c='b',alpha=0.7,label="squareline")
plt.plot(X,y_cube,c='r',alpha=0.5,label="cubeline")
plt.title("X vs Y")
plt.legend()
plt.show()
```

```python
categories =["apple","banana","mango","strawberry"]
values =[10,23,43,5]
plt.bar(categories,values,width=0.5,color=["red","blue","orange","yellow"],edgecolor='black')
plt.title("Counts of Fruits")
plt.ylabel("counts")
plt.xticks(rotation=45)
plt.show()
```

```python
categories =["apple","banana","mango","strawberry"]
values =[10,23,43,5]
plt.barh(categories,values,height=0.5,color=["red","blue","orange","yellow"],edgecolor='black')
plt.title("Counts of Fruits")
plt.ylabel("counts")
plt.xticks(rotation=45)
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt 
import numpy as np

data =  np.random.normal(loc=50,scale=10,size=1000)

plt.hist(data,bins=20,color='skyblue',edgecolor="black")

plt.xlabel('Vlaue')
plt.ylabel('Frequency')
plt.show()
```

# Scatterplots 



```python
from sklearn.datasets import make_blobs
X,y = make_blobs(
    n_features=2,n_samples=1000,random_state=42
)
print(X.shape)
```

```python
print(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis',edgecolors='black')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.title("make_blobs visualization")
plt.colorbar(label="cluster")
plt.show()
```
