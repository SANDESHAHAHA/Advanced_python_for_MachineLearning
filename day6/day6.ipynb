```python
aDict={}
with open("hello.txt",'r') as f:
    content=f.read().lower()
    words=content.split()
    for word in words:
        if word not in aDict:
            aDict[word]=1
        else:
            aDict[word]+=1
aDict
        
```

```python
with open("hello.txt",'r') as f:
    lines= f.readlines()
    for line in lines:
        print(line,end="")
```

```python
with open("new.txt",'w') as f:
    while(True):
        input_text=input("enter some text. press q or Q to quit")
        if input_text=='Q' | input_text=='q':
            break
        f.writelines(input_text)
        f.write("\n")
```

# reading csv file using csv module

```python
import csv
with open("file.csv",'r') as f:
    content=csv.reader(f)
    for rows in content:
        print(rows)
```

```python
with open("file.csv",'r') as f:
    content=csv.DictReader(f)
    for row in content:
        print(row['Name'],row['Address'],row['Age'])
```

```python
import csv 
data=[['Name','Address','Age'],
['John','Biratnagar',22],
['Hari','Itahari',32],
['SUman','Pokhara',23]
]
with open("new.csv",'w',newline="") as f:
    writer=csv.writer(f)
    writer.writerows(data)
    
```

```python
import csv
data=[
    {'Name':'John','Address':'Biratnagar','Age':12},
    {'Name':'Suman','Address':'Pokhara','Age':13},
    {'Name':'Hari','Address':'Itahari','Age':52},
]
with open("new1.csv",'w',newline="") as f:
    writer=csv.DictWriter(f,fieldnames=['Name','Address','Age'])
    writer.writeheader()
    writer.writerows(data)
```
