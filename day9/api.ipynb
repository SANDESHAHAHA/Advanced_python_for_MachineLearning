# SOAP APIS 
SOAP= Simple Object Access Protocal  \n
Uses Xml for requests and responses 
It does not requires url 
Requires a WSDL ( Web Services Description Language) URL 
WSDL is the language that acts as a middleware between two languages for ex: python and js 

```python
#conda install zeep library 
#wsdl is used for interconnection between banks 
from zeep import Client 
wsdl ="http://www.dnenline.com/calculator.asmx?WSDL"
client = Client(wsdl=wsdl)

# call a soap method 

result = client.service.Add(10,20)
print(result)
```
