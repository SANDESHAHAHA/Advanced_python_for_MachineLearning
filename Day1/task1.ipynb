```python
class Point: 
    def __init__(self,x,y):
        self.x = x
        self.y =y  
    
    def __str__(self):
        return f'({self.x},{self.y})'
    
    def __add__(self,other):
        # return (self.x + other.x , self.y+other.y)
        new_x =self.x+other.x
        new_y=self.y+other.y
        return Point(new_x,new_y)
p1=Point(1,2)
p2=Point(3,4)
p3=p1+p2 #p1.__add__(p2) under the hood 
print(p3)
        
```

```python
#wap to creat a class named distance that has feet and inches as data members . Overload any 2 operatros 
class Distance:
    def __init__(self,feet,inch):
        self.feet = feet
        self.inch = inch 
        
    def __add__(self,other):
        new_feet = self.feet+other.feet
        new_inch = self.inch+other.inch
        
        if new_inch>=12:
            new_feet=new_feet+int(new_inch/12)
            new_inch=new_inch%12
        return (new_feet,new_inch)
d1=Distance(1,12)
d2=Distance(2,13)
d3=d1+d2
print(d3)
            
            
            
   
```
