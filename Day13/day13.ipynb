#  Creating a Large dataset for chunking 

```python
import random 
import pandas as pd
import tqdm
```

```python
name =["Ram","Shyam","Hari","Sita","Suman"]
final_list =[]
for i in tqdm.tqdm(range(1,100001)):
    random_name = name[random.randint(0,4)]
    final_list.append({
        "EID":i,
        "Name":random_name,
        "Salary":100000*random.random()
    })
df = pd.DataFrame(final_list)
df.to_csv('big_data.csv', index=False)
```

```python
first_chunk = True 
output_file = 'processed_data.csv'
for chunk in pd.read_csv("big_data.csv",chunksize = 100000): # by default chunk size is 10k
    chunk["Gender"] = chunk["Name"].apply(lambda x: "Female" if x == "Sita" else "Male")
    chunk.to_csv(
        output_file,
        index=False,
        mode = "w" if first_chunk else "a",
        header=first_chunk
    )
```
